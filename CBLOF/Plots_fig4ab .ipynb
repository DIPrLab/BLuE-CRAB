```python
import numpy as np
import json
import matplotlib.pyplot as plt

with open('result_fig4a.json', 'r') as file:
    data = json.load(file)
keys_wdt = ['bledoubt_log_h.json', 'bledoubt_log_n.json','bledoubt_log_m.json']
scenarios=['W','D','T']
bc_f1sc = []
for k in keys_wdt:
    if k in data:
        bc_f1sc.append(data[k])
bd_f1sc=[0.9,0.8,0]


# Bar width and x locations
w, x = 0.4, np.arange(len(scenarios))

fig, ax = plt.subplots()
ax.bar(x - w/2, bc_f1sc, width=w, label='Ours', color='green',alpha=0.8,edgecolor='black')
ax.bar(x + w/2, bd_f1sc, width=w, label='BLE-Doubt Classifier',color='red',alpha=0.8,edgecolor='black')

ax.set_xticks(x)
ax.set_xticklabels(scenarios, fontsize=20,fontweight='bold')
ax.set_ylabel('F1 Score', fontsize=16,fontweight='bold')
# ax.set_xlabel('Scenario')
# ax.set_title('F1 Score')
ax.legend(loc='upper center',bbox_to_anchor=(0.5, 1.5), fancybox=True, shadow=True, ncol=2,frameon=False, fontsize=25)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.spines['bottom'].set_linewidth(0.5)
ax.tick_params(axis='x', labelsize=19)
ax.tick_params(axis='y', labelsize=17)
for label in ax.get_xticklabels():
    label.set_fontweight('bold')
for label in ax.get_yticklabels():
    label.set_fontweight('bold')
plt.tight_layout()
# plt.savefig('scenario2b.pdf',bbox_inches='tight')
plt.show()



```
